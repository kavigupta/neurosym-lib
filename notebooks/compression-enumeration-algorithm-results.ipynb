```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import itertools
import neurosym as ns
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("error")
```

```python
results = ns.examples.simple_dreamcoder.all_experiments("../outputs/simple_dreamcoder") 
```

```python
for compression_steps_by_iteration, count in results:
    result = results[compression_steps_by_iteration, count]
    timings, val_errors = [np.mean([r[k] for r in result], axis=0) for k in ["timings", "val_errors"]]
    timings = timings - timings[0]
    plt.plot(timings, val_errors, label=f"compress_steps={compression_steps_by_iteration}, enum_count={count}", marker=".")
plt.xlim(0, 300)
plt.xlabel("Time [s]")
plt.ylabel("MSE")
plt.legend()
plt.show()
```
